```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

```python
# gerando massa de dados
x, y = make_regression(n_samples=200, n_features=1, noise=30)
```

```python
# mostrando gráfico
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
# criação do modelo
modelo = LinearRegression()
```

```python
modelo.fit(x,y)
```

```python
modelo.intercept_ # coeficiente linear
```

```python
modelo.coef_ # coeficiente angular
```

```python
# mostrando resultado
import numpy as np
plt.scatter(x,y)
xreg = np.arange(-3, 3, 1)
plt.plot(xreg, modelo.coef_ * xreg - modelo.intercept_, color='red')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.30)
```

```python
modelo.fit(x_treino, y_treino)
```

```python
modelo.score(x_teste, y_teste) #r2
```
