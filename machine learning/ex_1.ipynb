```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate sample data
np.random.seed(0)
x1 = np.random.rand(100) * 10  # Independent variable 1
x2 = np.random.rand(100) * 10  # Independent variable 2
# Create a linear relationship with some noise
m1, m2, b = 3, 2, 5
y = m1 * x1 + m2 * x2 + b + np.random.randn(100) * 2  # Dependent variable

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of the data points
ax.scatter(x1, x2, y, color='blue', label='Data points')

# Create a meshgrid for the regression plane
x1_range = np.linspace(0, 10, 20)
x2_range = np.linspace(0, 10, 20)
x1_mesh, x2_mesh = np.meshgrid(x1_range, x2_range)

# Calculate the predicted y values for the regression plane
y_pred = m1 * x1_mesh + m2 * x2_mesh + b

# Plot the regression plane
ax.plot_surface(x1_mesh, x2_mesh, y_pred, alpha=0.5, color='orange', label='Regression Plane')

# Labels
ax.set_xlabel('X1 (Independent Variable 1)')
ax.set_ylabel('X2 (Independent Variable 2)')
ax.set_zlabel('Y (Dependent Variable)')
ax.set_title('Multiple Linear Regression Model')

plt.legend()
plt.show()
```

```python

from sklearn.linear_model import LinearRegression

# Dados de exemplo
data = {
    'Área': [50, 60, 70, 80, 90],
    'Preço': [150000, 180000, 200000, 230000, 250000]
}
df = pd.DataFrame(data)
print(df)

# Variáveis independentes e dependentes
X = df[['Área']]
y = df['Preço']

# Criando e ajustando o modelo
model = LinearRegression()
model.fit(X, y)

# Coeficientes
b = model.intercept_  # Coeficiente linear
m = model.coef_[0]    # Coeficiente angular

print(f"Coeficiente linear (b): {b}")
print(f"Coeficiente angular (m): {m}")
```

```python
dataframe = dataframe.drop(['id', 'date', 'long', 'lat', 'zipcode'], axis=1)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split

x = dataframe.drop('price', axis=1)
y = dataframe['price']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=14)
model = Ridge() # alpha=1
model_lin = LinearRegression()
model.fit(x_train, y_train)
model_lin.fit(x_train, y_train)
result = model.score(x_test, y_test)
print(result)
result = model_lin.score(x_test, y_test)
result
```
