```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', 21)
dataframe = pd.read_csv('data/kc_house_data.csv')
dataframe.head()
```

```python
dataframe = dataframe.drop(columns=['id', 'date', 'zipcode', 'lat', 'long'])
dataframe.tail()
```

```python
y = dataframe['price'] # target varieavle
x = dataframe.drop('price', axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=14)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
result = model.score(x_test, y_test) #r2
result
```

```python
from sklearn.linear_model import Ridge
model_ridge = Ridge()
model_ridge.fit(x_train, y_train)
result = model_ridge.score(x_test, y_test)
result
```
