```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
pd.set_option('display.max_columns', 30)
dataframe = load_breast_cancer()
x = pd.DataFrame(dataframe.data, columns=[dataframe.feature_names])    
y = pd.Series(dataframe.target)
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# definindo valoes que serão testados
values_c = np.array([0.01, 0.1, 0.5, 1, 2, 3, 5, 10, 20, 50, 100])
regulation = ['l1', 'l2']
grid_values = {
    'C':values_c,
    'penalty':regulation
}

# criando modelo
model = LogisticRegression()

# criando os grids
logistic_regration_grid = GridSearchCV(estimator=model, param_grid=grid_values, cv=5)
logistic_regration_grid.fit(x,y)

# imprimindo a melhor acurácia e os melhores parâmetros
print('Melhor acurácia: ', logistic_regration_grid.best_score_)
print('Parâmetro c: ', logistic_regration_grid.best_estimator_.C)
print('Regularização: ', logistic_regration_grid.best_estimator_.penalty)
```
