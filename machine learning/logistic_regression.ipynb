```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', 64)
pd.set_option('display.max_rows', 64)
dataframe = pd.read_csv('data/Data_train_reduced.csv')
dataframe.head()
```

```python
dataframe.shape
```

```python
dataframe.dtypes
```

```python
def percent_missing(dataframe):
    missing = dataframe.isnull().sum()
    return missing / len(dataframe['Instant.Liking']) * 100


percent_missing(dataframe)
```

```python
dataframe.columns
```

```python
dataframe.drop([
    'q8.20',
    'q8.18',
    'q8.17',
    'q8.10',
    'q8.8',
    'q8.9',
    'q8.2',
    'Respondent.ID',
    'Product',
    'q1_1.personal.opinion.of.this.Deodorant'
], axis=1, inplace=True)
dataframe.shape
```

```python
dataframe['q8.12'].fillna(dataframe['q8.12'].median(), inplace=True)
dataframe['q8.7'].fillna(dataframe['q8.7'].median(), inplace=True)
percent_missing(dataframe)
```

```python
y = dataframe['Instant.Liking']
x = dataframe.drop('Instant.Liking', axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, StratifiedGroupKFold

# separando os dados em folds
stratifiedkfold = StratifiedGroupKFold(n_splits=5)

# criando modelo
model = LogisticRegression()
result = cross_val_score(model, x, y, cv=stratifiedkfold)
result.mean()
```
